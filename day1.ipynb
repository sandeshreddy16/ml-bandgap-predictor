```python
import pandas as pd
data={ 
    "material": ["si","GaAs","Ti02","Fe","Cu"],
    "bandgap": [1.1,1.4,3.0,0.0,0.0],
    "is_metal": [False,False,False,True,True]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.shape)
print(df.describe()) # it will give different opertions 
print(df.dtypes) # it will tell types 
```

```python
#filtering concepts
metals=df[df["is_metal"] ==True]
semiconductors=df[df["bandgap"]>0.5]
print(semiconductors) # here it will  write the table contents
print(metals)
metals ## here it displays the table 
semiconductors
```

```python
# sorting 
print(df.sort_values("bandgap",ascending=False)) # it is based on the bandgap values and if ascending =true then it will be from smallest to largest now here it is false so it will be from largest to smallest
```

```python
# adding a column
df["bandgap_squared"] = df["bandgap"]**2
print(df)
```

```python
import matplotlib.pyplot as plt
#simple scatter plot
plt.figure(figsize=(7,4)) #width is 7 and height is 4 of the graph

#this below line(x axis and y axis ) and plt.show() is only important others are optional
plt.scatter(df["material"],df["bandgap"], # material will be in x axis and bandgap in y axis
            c=["red" if m else "blue" for m in df["is_metal"]], #like if m is true it wuill give red colour if not it will give blue colour
            s=100) # it is the pointer size means the dots size

plt.xlabel("Material")
plt.ylabel("Bandgap (ev)")
plt.title("Bandgap of common materials")
plt.axhline(0,color='gray',linestyle='--' ,alpha=0.5) # alpha means opacity 0 means invisible , 0.5 means semi transparent , 1 means solid transparent
plt.tight_layout() #it will adjust everything
plt.savefig("first_plot.png",dpi=150) ## dpi means dpi per inch 
plt.show()
print("plot saved")
          
```
