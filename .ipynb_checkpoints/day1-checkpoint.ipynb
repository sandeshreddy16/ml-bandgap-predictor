```python
import numpy as np 
a=np.array([1,2,3,4,5])
b=np.zeros((3,3))
c=np.ones((2,4))
d=np.full((2,4),3)
print(a.shape,b.shape,c.shape)
print(a)
print(b)
print(c)
print(d)
```

```python
print(a[1:4]) #slicing
print(a[a>1]) # filtering
```

```python
print(a[3]**2)
print(np.mean(a),np.std(a))
```

```python
n=np.array([[1,2],[3,4],[5,6]])
print(n.shape)
print(n[:,0]) # take all row and first column only
print(n[0,:]) # take all column and first row only
print(n.T) #transpose it will convert all row into columns and vice versa
```
