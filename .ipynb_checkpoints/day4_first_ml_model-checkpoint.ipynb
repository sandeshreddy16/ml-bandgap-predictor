```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# load pre-computed features — instant, no waiting
df = pd.read_csv("features_ready.csv")

# separate features and target
feature_cols = [c for c in df.columns 
                if c not in ['bandgap', 'formula', 'formation_energy']]
X = df[feature_cols].values
y = df['bandgap'].values

print(f"Loaded instantly — {df.shape[0]} materials, {len(feature_cols)} features")
print("Ready to train!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("materials_clean.csv")
print(f"loaded {len(df)} clean materials")
```

```python
from pymatgen.core import Composition
def to_composition(formula):
    try:
        return Composition(formula)
    except:
        return None

df['composition']=df['formula'].apply(to_composition) #the lhs side is the new column like if GaAs then in the column it gives Ga1 As1 
df=df.dropna(subset=['composition']) #dropna() removes rows containing missing values (NaN or None).
print(f"Compositions ready: {len(df)}")
```

```python
from matminer.featurizers.composition import ElementProperty
featurizer = ElementProperty.from_preset("magpie")
print(f"Features to generate: {len(featurizer.feature_labels())}")
print("Running... this takes a few minutes")
df=featurizer.featurize_dataframe(
    df,
    col_id='composition',
    ignore_errors=True)
print(f"Done , Shape: {df.shape}")
```

```python
feature_cols=featurizer.feature_labels() #get all generated feature names
X=df[feature_cols].copy()   #these are the input features and we have to write copy because if not written then it will give affect the df.
Y=df['bandgap'].copy()   #target variable  it selects the bandgap column
mask=X.notna().all(axis=1) # this finds rows without missing values x.notna() checks whether each value is not missing if missing then it gives False and .all(axis=1) means it checks each row and then it will stored in list which is mask
X=X[mask] #it removes row containing missing features
Y=Y[mask]  #it removes the corresponding band gap values too
print(f"X shape: {X.shape}") # it returns rows and columns.
print(f"Y range: {Y.min():.2f} eV") #it gives the smallest band gap till two decimal round off
X.to_csv("features_magpie.csv",index=False) #it creates a csv file which contains all the input features 
Y.to_csv("targets_bandgap.csv",index=False) # here index=false means do not add  the index in the csv file
print("saved")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.dummy import DummyRegressor
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
baseline = DummyRegressor(strategy='mean') # a dummy model which find the average band gap and always predict that
baseline.fit(X_train,Y_train) #generally model.fit() means learn patterns from data but dummy regressor does not learn patterns it only calculates the average of Y_train it takes the means of all the targets and stores it and it completely ignores the X_ train features
#the model now remembers the prediction value
mae_baseline=mean_absolute_error(Y_test,baseline.predict(X_test)) #now the Y_test gives the real target values and baseline.predict(X_test) gives the predicted value and then calculate the error caused by these two values (target and predicted ) and take the mean of them it gives the mean absolute error
#on average the model is wrong by the below value that is 0.6976
print(f"Baseline MAE: {mae_baseline:.4f} eV")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# fit() learns the mean and standard deviation from X_train
# transform() uses those values to scale X_train
# fit_transform() = fit() + transform()
X_train_scaled=scaler.fit_transform(X_train) #fit takes the features mean,std deviation and so much and stores in memory and transform() means it will apply this formula
#z=(x-mean_of_feature)/std deviation so by this it scale the training data
 
#Scale the test data using the SAME mean and std

# learned from the training data

# DO NOT use fit_transform(X_test)
# because fit() would learn information from the test set
# (called data leakage)

# transform() only applies the scaling rules already learned
# from X_train, which makes the evaluation fair
X_test_scaled=scaler.transform(X_test) #here we will not use fit() because already we got mean and std deviation from above 
lr=LinearRegression()
lr.fit(X_train_scaled,Y_train)
mae_lr=mean_absolute_error(Y_test,lr.predict(X_test_scaled))
# Use the trained Linear Regression model to predict band gaps
# for the test materials using their scaled features

# lr.predict(X_test_scaled) returns predicted band gap values

# Compare the predicted band gaps with the actual band gaps (Y_test)

# mean_absolute_error() calculates the average absolute difference
# between actual and predicted values

# Lower MAE = Better model performance
print(f"linear Regression MAE: {mae_lr:.4f} eV")
```

```python
# Linear Regression learns relationships between material features and band gap.
# Unlike DummyRegressor, it does not predict the same value for every material.
# If the features contain useful information, the MAE should be lower than the baseline MAE.
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(
    n_estimators=100,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1)
rf.fit(X_train,Y_train)
Y_pred_rf=rf.predict(X_test)
mae_rf=mean_absolute_error(Y_test,Y_pred_rf)
r2_rf=rf.score(X_test,Y_test) #r square means how much of the variation in band gap is explained by the model
print(f"Random Forest MAE: {mae_rf:.4f} eV")
print(f"Random Forest R²: {r2_rf:.4f}")
```

```python
# Import the Random Forest regression algorithm

# Create a Random Forest model

# Build 100 decision trees

# Each leaf node must contain at least 2 samples
# Helps reduce overfitting

# Fix randomness so results are reproducible

# Use all available CPU cores for faster training

# Train the model using training features (X_train)
# and training target values (Y_train)
# The model learns the relationship between
# material features and band gap

# Predict band gaps for unseen test materials
# using the learned patterns from training

# Calculate Mean Absolute Error (MAE)
# Compare actual band gaps (Y_test)
# with predicted band gaps (Y_pred_rf)
# MAE = average prediction error
# Lower MAE is better

# Calculate R² (R-squared) score
# Measures how well the model explains
# the variation in band gap values
# R² = 1.0 → perfect prediction
# R² = 0.0 → no better than predicting the mean
# R² < 0.0 → worse than predicting the mean

# Display the average prediction error

# Display the R² score of the model
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))
Y_test = Y_test.squeeze()
axes[0].scatter(Y_test,Y_pred_rf,alpha=0.4,s=15,color='#534AB7')  #alpha means transparency and s means size of each dot
lims=[min(Y_test.min(),Y_pred_rf.min()),max(Y_test.max(),Y_pred_rf.max())] # use to draw a refernce line
axes[0].plot(lims,lims,'r--',linewidth=1.5,label='Perfect')# x is lims and y is lims so y=x means predicted = actual ,r-- means red dashed label =perfect means name shown in legend
axes[0].set_xlabel('Actual Bandgap (eV)')
axes[0].set_ylabel('Predicted Bandgap (eV)')
axes[0].set_title(f'Random Forest - MAE = {mae_rf:.3f} eV')
axes[0].legend() # it displays the legend shows perfect nest to red dashed line 
#legend is a small box that explains what lines , colors ,or symbols in the graph represent.

errors=Y_pred_rf - Y_test

axes[1].hist(errors,bins=40,color='#1D9E75',edgecolor='white',alpha=0.85)# histogram graph , bins=40 means divide error range into 40 groups
axes[1].axvline(0,color='red',linestyle='--')#axvline means a verical line at error =0
axes[1].set_xlabel('Prediction Error (eV)')
axes[1].set_title('Error Distribution')
plt.tight_layout() # automatically adjusts spacing
plt.savefig('day4_rf_result.png',dpi =150) #dpi means image quality , dots per inch 
plt.show() #display the plot on screen 
```

```python
importances=pd.Series(
    rf.feature_importances_,index=feature_cols).sort_values(ascending=False)
plt.figure(figsize=(10,6))
importances.head(15).plot(kind='barh',color='#534AB7',alpha=0.85)
plt.xlabel('Feature Importance')
plt.title('Top 15 Features - Random Forest')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('day4_feature_importance.png',dpi=150)
plt.show()
print("Top 5 features:")
print(importances.head(5))
```

```python
from sklearn.model_selection import cross_val_score
cv_scores=cross_val_score(
    RandomForestRegressor(n_estimators=100,random_state=42,n_jobs=-1),
    X,Y,cv=5,
    scoring='neg_mean_absolute_error'
)
cv_mae=-cv_scores
print(f"MAE per fold: {cv_mae.round(4)}")
print(f"\n>>> REPORTABLE RESULT : MAE ={cv_mae.mean():.3f} ± {cv_mae.std():.3f} eV <<<")
```

```python
results=pd.DataFrame({
    'Model': ['Baseline','Linear Regression','Random Forest (5-CV)'],
    'MAE (eV)' : [round(mae_baseline,4),round(mae_lr,4),round(cv_mae.mean(),4)]
})
print(results.to_string(index=False))
results.to_csv("day4_results.csv",index=False)
```
